```python
from openai import OpenAI
from dotenv import load_dotenv
import time

client =OpenAI()

```

```python
ss = time.perf_counter()
small_prompt = client.responses.create(
    model = "gpt-5-nano",
    input ="Create a business idea to sell kids clothes sourced from india"
)
se = time.perf_counter()
small_prompt_latency = se - ss

```

```python
ms = time.perf_counter()
medium_prompt = client.responses.create(
    model = "gpt-5-nano",
    input ="Create a business idea to sell kids clothes sourced from india in US with low inventory and less money to invest. step by step "
)
me = time.perf_counter()
medium_prompt_latency = me - ms
```

```python
ls = time.perf_counter()
long_prompt = client.responses.create(
    model = "gpt-5-nano",
    input ="""""
    Create a business idea to sell kids clothes sourced from india in US
    Constraints-
        1. minimal investment, less than $10000. 
        2. shipping from india in a bulk with set buy. that means if selected one piece it will have all the sizes. 
        3.  baby and toddler only
    Goal: to create a successful business with shipping first from US(by buying in bulk) so that the shipping time is reduced later, direct shipping from india to control inventory and waste

    Given: already have manufacturer contacts 
    Recommend if this business is feasible in US. if yes, how to create the business step by step 

    
    """""
)
le = time.perf_counter()
long_prompt_latency = le - ls
```

```python
vls = time.perf_counter()
vlong_prompt = client.responses.create(
    model = "gpt-5-nano",
    input ="""""
    Create a business idea to sell kids clothes sourced from india in US
    Constraints-
        1. minimal investment, less than $10000. 
        2. shipping from india in a bulk with set buy. that means if selected one piece it will have all the sizes. 
        3.  baby and toddler only
        4. i dont want to spend lot of money
    Goal: to create a successful business with shipping first from US(by buying in bulk) so that the shipping time is reduced later, direct shipping from india to control inventory and waste
    how can i make it profitable. what about markrting strategy? what about seo? what about entering into the market? what about making sure that there is sales every month? what about ads? how do it position the ads properly? should i make a website of my own or sall it by amazon or etsy channnels? please give full detail answers. make sure that there is no mistake
    Given: already have manufacturer contacts 
    Recommend if this business is feasible in US. if yes, how to create the business step by step 

    
    """""
)
vle = time.perf_counter()
vlong_prompt_latency = vle - vls
```

```python
results = [

    {
        "prompt_type":"small",
        "input_tokens":small_prompt.usage.input_tokens,
        "output_tokens": small_prompt.usage.output_tokens,
        "latency": small_prompt_latency
    },
    {
        "prompt_type":"medium",
        "input_tokens":medium_prompt.usage.input_tokens,
        "output_tokens": medium_prompt.usage.output_tokens,
        "latency": medium_prompt_latency
    },
     {
        "prompt_type":"long",
        "input_tokens":long_prompt.usage.input_tokens,
        "output_tokens": long_prompt.usage.output_tokens,
        "latency": long_prompt_latency
    },
    {
        "prompt_type":"vlong",
        "input_tokens":vlong_prompt.usage.input_tokens,
        "output_tokens": vlong_prompt.usage.output_tokens,
        "latency": vlong_prompt_latency
    }
]
```
